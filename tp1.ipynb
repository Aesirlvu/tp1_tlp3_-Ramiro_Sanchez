### 2. **Operaciones Matemáticas Básicas**

### 3. **Condicionales (if-else)**

   - **Actividad 1.1**: Escribir un programa que pida al usuario su edad y le diga si es mayor de edad (18 años o más).
   - **Actividad 1.2**: Agregar un mensaje diferente para diferentes rangos de edad, como por ejemplo, si es un niño, adolescente o adulto.

```python

# todo: Resolución

human_ages = list(range(0, 122))

kid_age_range = [age for age in human_ages if age <= 12]
teenager_range = [age for age in human_ages if age >= 13 and age <= 19]
adult_age_range = [age for age in human_ages if age >= 20 and age <= 59]
elder_age_range = [age for age in human_ages if age >= 60]


max_kid_age = 12

min_teenager_age = 13

legal_age = 18

max_teenager_age = 19

min_adult_age = 20
max_adult_age = 59

min_elder_age = 60


# --- ENTRADA DE DATOS ---

age = int(input("-- Ingresa tu edad: "))

def isHuman(age):
    if age in human_ages:
        return True
    else:
        print("cómo estás vivo!?")
        return False


def legal_age_validator(age):
    if age >= legal_age:
        print(f"legalmente mayor de edad, ya tienes: {age}!")
        return True
    else:
        return False

def age_classifier(age):
    if age <= max_kid_age:
        return print(f"aún eres un niño, tienes: {age} años!")
    elif min_teenager_age <= age <= max_teenager_age:
        return print(f"adolescente de {age} años!")
    elif min_adult_age <= age <= max_adult_age:
        return print(f"adulto de {age} años!")
    elif age >= min_elder_age:
        return print(f"persona mayor de {age} años!")

if isHuman(age):
    legal_age_validator(age)
    age_classifier(age)
```

### 10. **Proyecto Final: Calculadora**

   - **Actividad 1.1**: Los estudiantes deben crear una calculadora básica que permita realizar operaciones de suma, resta, multiplicación y división con entradas del usuario. El programa debe pedir al usuario que elija la operación y luego introducir 2 números para realizar el cálculo.
   - **Actividad 1.2**: Implementar una opción para que el usuario pueda continuar haciendo cálculos o salir del programa.

```python
# todo: Resolución

def suma(num1, num2):
    return num1 + num2

def resta(num1, num2):
    return num1 - num2

def multiplicacion(num1, num2):
    return num1 * num2

def division(num1, num2):
    try:
        return num1 / num2
    except ZeroDivisionError:
        return "No se puede dividir por cero."

while True:
    print("\nOperaciones disponibles:")
    print("1. Suma")
    print("2. Resta")
    print("3. Multiplicación")
    print("4. División")
    print("5. Salir")

    opcion = input("Selecciona una operación (1-5): ")

    if opcion == "5":
        print("Saliendo del programa. ¡Hasta luego!")
        break

    if opcion in ["1", "2", "3", "4"]:
        try:
            num1 = float(input("Ingresa el primer número: "))
            num2 = float(input("Ingresa el segundo número: "))
            if opcion == "1":
                print(f"Resultado: {suma(num1, num2)}")
            elif opcion == "2":
                print(f"Resultado: {resta(num1, num2)}")
            elif opcion == "3":
                print(f"Resultado: {multiplicacion(num1, num2)}")
            elif opcion == "4":
                print(f"Resultado: {division(num1, num2)}")
        except ValueError:
            print("ingresa valores numéricos válidos")
    else:
        print("opción no válida, selecciona una opción entre 1 y 5.")
```

### EXTRA. **Juegos Simples (Adivina el número)**

   - **Actividad 1.1**: Escribir un juego donde el programa elija un número aleatorio entre 1 y 100 y el usuario tenga que adivinarlo. El programa debe dar pistas si el número es mayor o menor.
   - **Actividad 1.2**: Limitar el número de intentos y mostrar un mensaje de éxito o fracaso al final.

```python
numero_ganador = 30

intentos = 0
max_intentos = 3

while intentos <= max_intentos:
    numero_ingresado = int(input("-- ingresa un número (entre 1 al 100): "))

    intentos_restantes = max_intentos - intentos

    print(f"intentos restantes: {intentos_restantes}")

    if numero_ingresado == numero_ganador:
        print(f"Felicidades, el numero ganador era: {numero_ganador}")
        break
    if numero_ingresado > numero_ganador:
        print("El número es menor!")
    elif numero_ingresado < numero_ganador:
        print("el numero es mayor!")
    else:
        print("Ese no parece ser el número ganador..!")
    intentos += 1

    if intentos == max_intentos:
        print("Haz alcanzado el límite de intentos!")
        break
```
